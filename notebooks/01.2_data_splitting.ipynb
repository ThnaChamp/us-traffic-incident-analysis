# US Accidents - Data Splitting
ใน Notebook นี้เราจะดำเนินการแบ่งข้อมูล (Data Splitting) ออกเป็น **Train Set** และ **Test Set** เพื่อใช้ในขั้นตอนการทำ Advanced Cleaning, Feature Engineering และ Training ต่อไป

## 1. Setup and Imports

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

## 2. Load Basic Cleaned Data and Perform Split
เราจะแบ่งข้อมูลในอัตราส่วน 80:20 (Train:Test) โดยใช้ `random_state=42` เพื่อให้ผลลัพธ์คงที่

```python
df = pd.read_csv("../data/processed/01.1/accidents_basic_clean.csv")

save_dir = "../data/processed/01.2"

df_train, df_test = train_test_split(df, test_size=0.2, random_state=42)

print(f"Number of df_train rows: {df_train.shape[0]:,} rows")
print(f"Number of df_test rows: {df_test.shape[0]:,} rows")
```

## 3. Export Split Data
บันทึกข้อมูลออกเป็น 3 ส่วน:
1. **train_data.csv**: ข้อมูลสำหรับใช้ Training และ Advanced Cleaning
2. **test_data.csv**: Features ของ Test Set (ตัดเฉลยออก)
3. **test_secret_answers.csv**: เฉลย (Ground Truth) ของ Test Set สำหรับใช้ในขั้นตอน Evaluation

```python
df_train.to_csv(f"{save_dir}/train_data.csv", index=False)

test_features = df_test.drop(columns=["Duration(min)","Distance(mi)"])
test_features.to_csv(f"{save_dir}/test_data.csv", index=False)

test_answers = df_test[["Duration(min)", "Distance(mi)"]] 
test_answers.to_csv(f"{save_dir}/test_secret_answers.csv", index=False)
```
