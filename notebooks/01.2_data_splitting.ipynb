```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("../data/processed/01.1/accidents_basic_clean.csv")

save_dir = "../data/processed/01.2"

df_train, df_test = train_test_split(df, test_size=0.2, random_state=42)

print(f"Number of df_train rows: {df_train.shape[0]:,} rows")
print(f"Number of df_test rows: {df_test.shape[0]:,} rows")
```

```python
df_train.to_csv(f"{save_dir}/train_data.csv", index=False)

test_features = df_test.drop(columns=["Duration(min)","Distance(mi)"])
test_features.to_csv(f"{save_dir}/test_data.csv", index=False)

test_answers = df_test[["Duration(min)", "Distance(mi)"]] 
test_answers.to_csv(f"{save_dir}/test_secret_answers.csv", index=False)
```
